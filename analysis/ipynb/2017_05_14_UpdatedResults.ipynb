```python
def plot_sample_worker_PR(threshold_col_lst,y_axis='Precision',init_fig=True,color=''):
    if init_fig: plt.figure()
    y_err_lst = []
    y_val_lst =[]
    sample_lst = sorted(Nsample_lst)
    avrg_tbl = df_all.groupby("Nworker")[threshold_col_lst].mean()
    err_tbl = df_all.groupby("Nworker")[threshold_col_lst].std()
    for attr in threshold_col_lst:
        y_val_lst.append(list(avrg_tbl[attr]))
        y_err_lst.append(list(err_tbl[attr]))
    
    for i in range(len(y_val_lst)):
        if color=="":
            plt.plot(sample_lst,y_val_lst[i],label=threshold_col_lst[i])
        else:
            plt.plot(sample_lst,y_val_lst[i],label=threshold_col_lst[i],color=color)
        #plt.errorbar(sample_lst,y_val_lst[i],label=selected_col_lst[i], yerr=y_err_lst[i],capsize=3)
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
    plt.xlabel("N sample",fontsize=13)
    plt.ylabel(y_axis,fontsize=13)
#     plt.ylim(0.8,1)
#     plt.xlim(5,15)
```

# Effects of varying threshold

```python
algo = "PixelEM"
P_tbl = df_all.filter(regex=algo).filter(regex="P ").mean()
R_tbl = df_all.filter(regex=algo).filter(regex="R ").mean()
```

```python
Pthres_lst = [int(i.split('=')[-1][:-1]) for i in P_tbl.keys()]
Rthres_lst = [int(i.split('=')[-1][:-1]) for i in R_tbl.keys()]
```

```python
df_all.filter(regex="TileEM").filter(regex="P ").mean()
```

```python
def plot_vary_threshold(Nworker,algo = "PixelEM"):
    plt.figure()
    df = df_all[df_all["Nworker"]==Nworker]
    P_tbl = df.filter(regex=algo).filter(regex="P ").mean()
    R_tbl = df.filter(regex=algo).filter(regex="R ").mean()
    thres_lst = list(set([int(i.split('=')[-1][:-1]) for i in P_tbl.keys()]))
    order = np.argsort(thres_lst)
    ordered_thres_lst = np.array(thres_lst)[order]
    print ordered_thres_lst
    plt.plot(ordered_thres_lst,np.array(P_tbl)[order],label="Precision")
    plt.plot(ordered_thres_lst,np.array(R_tbl)[order],label="Recall")
    plt.title("{}  N worker={}".format(algo,Nworker),fontsize=14)
    plt.xlabel("Threshold",fontsize=13)
    plt.legend()
```

```python
for Nworker in worker_Nbatches.keys():
    plot_vary_threshold(Nworker,algo = "PixelEM")
    plot_vary_threshold(Nworker,algo = "TileEM")
    plot_vary_threshold(Nworker,algo = "GT PixelEM")
```
